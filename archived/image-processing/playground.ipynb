```python
import pathlib
import skimage
from matplotlib import pyplot as plt
from matplotlib import image as mpimage

def show(im):
    plt.imshow(im, cmap="gray")
    plt.axis("off")
    plt.show()
```

```python
from skimage import filters
from skimage import color

# im = skimage.io.imread(pathlib.Path('IMG_3909.JPG'))
im = skimage.io.imread(pathlib.Path('_MG_6213.CR2.jpg'))
im_gray = color.rgb2gray(im)
# im2 = im2 < 0.006
```

```python
threshold = filters.try_all_threshold(im_gray)
# threshold = filters.threshold_local(im2, block_size=21, offset=0.000001)
im2 = im_gray > threshold
show(im2)
```

```python
from skimage import exposure

im3 = exposure.equalize_adapthist(im_gray)
# im3 = exposure.adjust_gamma(im3, gamma=0.23)
show(im3)
```
