```python
import numpy as np
from matplotlib import pyplot as plt
import os
from contextlib import contextmanager
from typing import Literal
import pathlib
from astropy.io import fits
from astropy import visualization

Channel = Literal['L'] | Literal['R'] | Literal['G'] | Literal['B']

class Input:
    def __init__(self, base_directory: pathlib.Path):
        self.base_directory = base_directory

    def luminosity(self):
        return self.base_directory / 'L'
        
def display_fits(fits):
    data = fits[0].data
    stretch = visualization.AsinhStretch(a=0.01)
    visualization.imshow_norm(data, stretch=stretch, cmap="gray")
    plt.axis('off')
    plt.show()

input = Input(pathlib.Path('../input/NGC 1365 Great Barred Galaxy'))
for file in os.listdir(input.luminosity())[0:2]:
    with fits.open(input.luminosity() / file) as f:
        display_fits(f)
```
